#Ejercicios

1. Buscar todos los libros en castellano que contengan la palabra Normandía en el título

```
filtrado1 = (df['titulo'].str.contains('Normandía')) & (df['idioma'] == 'Español')

ejercicio1 = df.loc[filtrado1]

ejercicio1
```
